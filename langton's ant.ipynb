```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def get_next_move(i, j, next_direction):
    
    if next_direction == 0:
        j-=1
    elif next_direction == 1:
        i+=1
    elif next_direction == 2:
        j+=1
    else:
        i-=1
    return i,j
```

```python
def langton_ant_general(n: int, states, directions, max_iters=10000, skip=1000) -> None:

    grid = np.zeros(shape=[n,n])
    i,j = n//2, n//2
    current_direction = states[0]

    for epoch in range(max_iters):
        #exit if left grid
        if i<0 or j<0 or i>n-1 or j>n-1:
            break
        # get turmite direction
        current_state = int(grid[i,j])
        turn = directions[current_state]
        if turn=='R':
            current_direction = (current_direction+1)%4
        else:
            current_direction = (current_direction-1)%4
        
        #change state
        grid[i,j] = (grid[i,j]+1) % len(directions)
        
        #move
        i,j = get_next_move(i,j,current_direction)
        
        
        
        #print
        if epoch%skip== 0:
            plt.imshow(grid, "twilight")
            plt.show()
```

```python
directions = 'RRLLLRRL'
states  = np.arange(len(directions))
langton_ant_general(120, states, directions, 140000, skip=140000-1)
```

```python
directions = 'LRRRRRLLR'
states  = np.arange(len(directions))
langton_ant_general(120, states, directions, 140000, skip=10000)
```
